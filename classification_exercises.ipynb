```python
def split_function(df: pd.DataFrame, target_varible: str) -> int:
    train, test = train_test_split(df,
                                   random_state=123,
                                   test_size=.20,
                                   stratify= df[target_varible])
    
    train, validate = train_test_split(train,
                                   random_state=123,
                                   test_size=.25,
                                   stratify= train[target_varible])
    return train, validate, test
```

```python
# Inital take aways
# month to month is the most popular plan 
# costomers are more likly to stay if they are signed up for a year or two year plan 

# categorical v categorical sitch
# which means that a chi-squared test would be valuable
```

```python
# null: contract type and churn are indepdent of eachother
# alt: contract type and chrun have a dependent distribution

alpha = 0.05

observed = pd.crosstab(train_telco.contract_type, train_telco.churn)
observed
```

```python
chi2, p, degf, expected = stats.chi2_contingency(observed)
print(f'Chi2 -> {chi2}')
print(f'P-value -> {p}')
```

```python
if p < alpha:
    print('we reject the null hypothesis')
else:
    print('we fail to reject the null hypothesis')
```
